## Кодировнаие категориальных признаков

```python
cotegorial_cols = df.select_dtypes(include=['object']).columns
print(cotegorial_cols)
df_encoded = pd.get_dummies(df, columns=cotegorial_cols, drop_first=True)
print(df_encoded)
```

## Визуализация данных

```python
sns.set_style('whitegrid')
sns.set(font_scale=1.2)
```

```python
# гистограммы числовых признаков 

fig, axes = plt.subplots(3, 2, figsize=(15, 12))
axes = axes.flatten()

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

for i, col in enumerate(numeric_cols[1:]):
    sns.histplot(df[col], kde=True, ax=axes[i])
    axes[i].set(ylabel=None)  # Аналог set_ylabel(''), но короче
    #sns.despine(ax=axes[i])  # Чистим лишние линии (опционально)
plt.tight_layout()

plt.show()
```

```python
# Матрица корреляций
plt.figure (figsize=(15,12))
correlation_matrix = df.select_dtypes(include=['float64', 'int64']).corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Матрица корреляций')
plt.show()
```

```python
# Диаграмма рессеяния
sns.pairplot(df[numeric_cols[1:]])
plt.show()
```

## Обработка пропущенных значений

```python
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
#замена Age
value = df['Age'].median()
print(value)
df['Age'] = df['Age'].fillna(value)

#замена Embarked
df['Embarked'] = df ['Embarked'].fillna('S')

print(df.isnull().sum())
```
